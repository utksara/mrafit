```python
import mrafit.wavelet_bases as wavelet_bases
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error as mae

def sigmoid(x):
    return 1/(1 + np.exp(-x))

X = np.linspace(-5, 5, 500)
basis = wavelet_bases.GaussPleteauBasis(resolution=2)
coeffs, Yet, err = basis.get_mra_approx(sigmoid, X)
plt.plot(X, Yet)
plt.plot(X, [sigmoid(x) for x in X])
```

```python
def msig(x):
    return np.exp(-x**2)* (x**2 - x + 1)
    
N = 2000
resolution = 0.05
X = np.linspace(-5, 5, N)
basis = wavelet_bases.GaussletBasis(resolution=resolution)
coeffs, Yapprx, err = basis.get_mra_approx(msig, X, plotting_enabled=False)
# plt.plot(X, basis.get_wavefunction_values(0, X))
plt.plot(X, Yapprx, label="approx")
plt.plot(X, [msig(x) for x in X], label="actual")
plt.legend()
```
